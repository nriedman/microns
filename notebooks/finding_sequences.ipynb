```python
import numpy as np
import pandas as pd
import numpy as np
import math
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
```

```python
# Specify k
k = 5
```

```python
# Import k-dependent data
clusters_base = pd.read_csv(f'data/synapse_clusters/clusters_k{k}.csv', index_col=0)
clusters_w_pos = pd.read_csv(f'data/synapse_clusters/with_syn_ctrs/clusters_k{k}.csv', index_col=0)
clusters_w_ids = pd.read_csv(f'data/synapse_clusters/with_pre_ids/clusters_k{k}.csv', index_col=0)
print(clusters_w_pos.shape, clusters_w_ids.shape, clusters_base.shape)
display(clusters_base.head())
display(clusters_w_pos.head())
display(clusters_w_ids.head())
```

```python
# Slim down dataframes to only include relevant columns:
dims = ['x', 'y', 'z']
pos_keys = [f'synapse_ctr_{dim}_{i}' for i in range(k) for dim in dims ]
clusters_pos = clusters_w_pos[pos_keys]
display(clusters_pos.head())
```

```python
id_keys = [f'pre_pt_root_id_{i}' for i in range(k)]
clusters_ids = clusters_w_ids[id_keys]
display(clusters_ids.head())
```

```python
def dist_row_to_sequence(row):
    sequence = []
    for i in range(len(row)):
        next_index = np.argmin(row)
        sequence.append(next_index)
        row[next_index] = np.inf
    return sequence
```

```python
def get_id(index, pre_ids):
    return pre_ids.loc[f'pre_pt_root_id_{index}']

def index_to_id(index_sequence, pre_ids):
    id_sequence = []
    for index in index_sequence:
        id_sequence.append(get_id(index, pre_ids))
    return id_sequence
```

```python
# Given a cluster, find the sequence of synapses
def calculate_sequence(synapse_pos, pre_ids, k):
    synapse_pos = synapse_pos.to_numpy().reshape(k,3)
    pairwise_distances = squareform(pdist(synapse_pos))

    # Find the furthest points
    furthest_condensed = np.argmax(pairwise_distances)
    start,finish = np.unravel_index(furthest_condensed, (k,k))
    
    # Find the sequence of indexes
    sequence_row = pairwise_distances[start,:]
    sequence_indexes = dist_row_to_sequence(sequence_row)

    # Convert to sequence of ids
    sequence = index_to_id(sequence_indexes, pre_ids)

    return sequence
```

```python
def find_sequences(cluster_pos, cluster_ids, k):
    sequences = []
    for cluster_id, synapse_pos in cluster_pos.iterrows():
        pre_ids = cluster_ids.loc[cluster_id,:]
        sequence = calculate_sequence(synapse_pos, pre_ids, k)
        sequences.append(sequence)
    return sequences
```

```python
sequence_keys = [f'sequence_rank_{i}' for i in range(k)]
sequences = find_sequences(clusters_pos, clusters_ids, k)
```

```python
clusters_sequences = clusters_base.copy()
clusters_sequences.loc[:,sequence_keys] = sequences
display(clusters_sequences.head())
```

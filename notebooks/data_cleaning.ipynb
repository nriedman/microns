```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

```python
old_synapses = pd.read_csv('deprecated/synapses_no_repeats.csv', index_col=[0])
print(old_synapses.head())
```

```python
old_synapses[['pt_x_pre', 'pt_y_pre', 'pt_x_post', 'pt_y_post']] *= 4 / 1000
old_synapses[['pt_z_pre', 'pt_z_post']] *= 40 / 1000
print(old_synapses.head())
```

```python
ctr_pos = old_synapses['ctr_pt_position'].str.strip('[]').str.split(expand=True).astype(int)
ctr_pos.columns = ['ctr_pt_x', 'ctr_pt_y', 'ctr_pt_z']
old_synapses = pd.concat([old_synapses, ctr_pos], axis=1)
print(old_synapses.head())
```

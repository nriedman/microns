```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# First: Remove the cells that were repeated in the original dataset
# Load the data
old_cells = pd.read_csv('data/raw/exc_cells.csv')
cells_no_repeats = pd.read_csv('data/cells_split_xyz_no_duplicates.csv')
old_cells_pruned = old_cells.drop_duplicates(subset=['pt_root_id'])
print(old_cells.shape)
print(cells_no_repeats.shape)
print(old_cells_pruned.shape)
```

```python
old_cells_by_id = old_cells.groupby('pt_root_id')
cell_repeat_counts = old_cells_by_id.size()
repeated_cell_ids = cell_repeat_counts[cell_repeat_counts > 1].index.tolist()
display(repeated_cell_ids)
```

```python
new_cells_pruned = cells_no_repeats[~cells_no_repeats['pt_root_id'].isin(repeated_cell_ids)]
print(new_cells_pruned.shape)
print(len(new_cells_pruned) - len(old_cells_pruned))
```

```python
new_cells_pruned.to_csv('data/cells_no_repeats.csv', index=False)
```

```python
synapses = pd.read_csv('data/combined_synapse_data_rescaled.csv', index_col=[0])
print(synapses.shape)
```

```python
synapses_pruned = synapses[synapses['pre_pt_root_id'].isin(new_cells_pruned['pt_root_id']) &
                           synapses['post_pt_root_id'].isin(new_cells_pruned['pt_root_id'])]
print(synapses_pruned.shape)
```

```python
synapses_pruned.to_csv('data/synapses_no_repeats.csv')
```

```python
print((len(synapses) - len(synapses_pruned)) / len(synapses))
```

```python
synapses_repeated = synapses[synapses['pre_pt_root_id'].isin(repeated_cell_ids) |
                             synapses['post_pt_root_id'].isin(repeated_cell_ids)]
print(synapses_repeated.shape)
```
